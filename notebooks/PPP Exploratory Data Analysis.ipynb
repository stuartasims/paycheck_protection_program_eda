# Exploring Colorado's Paycheck Protection Program Loan Data

The federal government just released some data on the Paycheck Protection Program Loan recipients. For loans above 150,000, the name, type, address, and some details about ownership were released along with binned ranges for loan amounts. Unfortunately, they did not include specific dollar amounts for loans in this range.

For loans up to $150,000 they did release the exact loan amount but not the business name or specific address. Take a look at some of my data exploration below.

One note: Before well all go hunting down businesses abusing the program and making big claims, some businesses in this data have stated they did not receive, or even apply for, loans under this program but are in the dataset .See [this article](http://mail01.tinyletterapp.com/data-is-plural/data-is-plural-2020-07-08-edition/17448558-www.cnbc.com/2020/07/06/ppp-small-business-coroniavirus-loan-database-contains-errors.html?c=edd6cfc9-960d-4b6c-a66f-c72077ac5ae2) for some coverage. Until we know the veracity of these claims, it's probably best not to go on the hunt for recipients  that abused the program. That being said, [it's obvious and increasingly reported that some recipients don't fit the spirit of a "Small Business" by any stretch of the imagination and should raise some questions](https://chicago.cbslocal.com/2020/07/06/kanye-wests-yeezy-got-millions-from-federal-governments-paycheck-protection-program/).


**Note that the data in this notebook is exclusive to Loans where the recorded state is Colorado. High-Level takeaways are in blue box callouts.**
